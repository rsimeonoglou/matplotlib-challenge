## Observations and Insights 

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np
# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single dataset
mouse_df = pd.merge(mouse_metadata,study_results,on="Mouse ID")
# Display the data table for preview
mouse_df
```

```python
# Checking the number of mice.
mouse_df["Mouse ID"].nunique()
```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
duplicated_mouse_df= mouse_df[mouse_df.duplicated(["Mouse ID","Timepoint"])]
duplicated_mouse_df
```

```python
# Optional: Get all the data for the duplicate mouse ID. 

```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
mouse_df = mouse_df[mouse_df["Mouse ID"] != "g989"]
mouse_df
```

```python
# Checking the number of mice in the clean DataFrame.
mouse_df["Mouse ID"].nunique()
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
# Assemble the resulting series into a single summary dataframe.
regimen_group = mouse_df.groupby("Drug Regimen")
mean_tumor = regimen_group["Tumor Volume (mm3)"].mean()
median_tumor = regimen_group["Tumor Volume (mm3)"].median() 
var_tumor = regimen_group["Tumor Volume (mm3)"].var()
std_tumor = regimen_group["Tumor Volume (mm3)"].std()
sem_tumor = regimen_group["Tumor Volume (mm3)"].sem()

tumor_by_regimen = pd.DataFrame({"Mean":mean_tumor,
                                "Median":median_tumor,
                                "Variance":var_tumor,
                                "Standard Deviation":std_tumor,
                                "Standard Errors (SEM)":sem_tumor})
tumor_by_regimen
```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Using the aggregation method, produce the same summary statistics in a single line
tumor_by_regimen2 = mouse_df.groupby("Drug Regimen").agg({"Tumor Volume (mm3)":["mean","median","var","std","sem"]})
tumor_by_regimen2

```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin

# Start by getting the last (greatest) timepoint for each mouse
promising_regimen_df = mouse_df.loc[(mouse_df["Drug Regimen"]=="Capomulin")|
                                        (mouse_df["Drug Regimen"]=="Ramicane") |
                                        (mouse_df["Drug Regimen"]=="Infubinol")|
                                        (mouse_df["Drug Regimen"]=="Ceftamin")]
mouse_groupby = promising_regimen_df.groupby("Mouse ID")

max_timepoint = mouse_groupby["Timepoint"].max()
max_timepoint = max_timepoint.reset_index()

# Merge this group df with the original dataframe to get the tumor volume at the last timepoint
promising_regimen_df = pd.merge(max_timepoint,promising_regimen_df,on=["Mouse ID","Timepoint"])
promising_regimen_df
```

```python
# Put treatments into a list for for loop (and later for plot labels)

treatments = promising_regimen_df["Drug Regimen"].unique().tolist()

# Create empty list to fill with tumor vol data (for plotting)

tumor_vol = []

for treatment in treatments:
    treatment_df = promising_regimen_df.loc[(promising_regimen_df["Drug Regimen"]==treatment)]
    each_treatment_tumors = treatment_df["Tumor Volume (mm3)"]
    tumor_vol.append(each_treatment_tumors)

# Calculate the IQR and quantitatively determine if there are any potential outliers.    
    quartiles = each_treatment_tumors.quantile([.25,.5,.75])
    lowerq = quartiles[0.25]
    upperq = quartiles[0.75]
    iqr = upperq-lowerq
    print(f'{treatment}:')
    print(f"The lower quartile of {treatment} is: {lowerq}")
    print(f"The upper quartile of {treatment} is: {upperq}")
    print(f"The interquartile range of {treatment} is: {iqr}")
    print(f"The the median of {treatment} is: {quartiles[0.50]} ")
    lower_bound = lowerq - (1.5*iqr)
    upper_bound = upperq + (1.5*iqr)
    print(f"Values below {lower_bound} could be outliers for {treatment}.")
    print(f"Values above {upper_bound} could be outliers for {treatment}.")
    print(f"---------------------------------------------------------")
# Determine outliers using upper and lower bounds
    outlier_occupancy = treatment_df.loc[(treatment_df['Tumor Volume (mm3)'] < lower_bound) | (treatment_df['Tumor Volume (mm3)'] > upper_bound)]
    outlier_count = outlier_occupancy["Mouse ID"].count()
    if outlier_count > 0:
        print(f'The amount of outliers for {treatment} is {outlier_count}.')
        print(f'{outlier_occupancy[["Mouse ID","Tumor Volume (mm3)"]]}')
        
    else:
        print(f'There are no outliers for {treatment}')
        
    print(f"---------------------------------------------------------")
    

```

```python
# Generate a box plot of the final tumor volume of each mouse across four regimens of interest

red_square = dict(markerfacecolor='r', marker='s')
fig1, ax1 = plt.subplots()

ax1.set_title('Regimens of Interest')
ax1.set_ylabel('Tumor Volume')
ax1.set_xlabel("Regimens")

ax1.boxplot(tumor_vol, flierprops=red_square)

plt.xticks(np.arange(4)+1,treatments)

plt.show()

```

## Line and Scatter Plots

```python
# Generate a line plot of tumor volume vs. time point for a mouse treated with Capomulin
capomulin_mice = mouse_df.loc[(mouse_df["Drug Regimen"]=="Capomulin")]
mouse_sample = capomulin_mice["Mouse ID"].sample(1).reset_index()
mouse_sample = mouse_sample.iloc[0]['Mouse ID']

random_mouse_df = capomulin_mice.loc[(capomulin_mice["Mouse ID"]== mouse_sample )]
random_mouse_df
print(f"Mouse: {mouse_sample}'s Tumor size by Timepoint")
plt.figure(figsize=(10,5))
plt.title("Sample Mouse's Tumor Volume over Time")
plt.xlabel("Timepoint")
plt.ylabel("Tumor Volume")
sample_mouse_plot = plt.plot(random_mouse_df['Timepoint'],
                             random_mouse_df['Tumor Volume (mm3)'],
                             linewidth=3.0,
                             color="blue")

```

```python
# Generate a scatter plot of average tumor volume vs. mouse weight for the Capomulin regimen
capomulin_mice_groupby = capomulin_mice.groupby("Mouse ID")
avg_tumor = capomulin_mice_groupby["Tumor Volume (mm3)"].mean().reset_index()
weight = capomulin_mice.drop_duplicates(subset=['Mouse ID'])
weight = weight[["Mouse ID","Weight (g)"]]

tumor_by_weight = pd.merge(avg_tumor,weight,on=["Mouse ID"])

scatterplot = plt.scatter(tumor_by_weight["Weight (g)"],tumor_by_weight["Tumor Volume (mm3)"])
plt.xlabel('Weight (g)')
plt.ylabel('Tumor Volume (mm3)')
plt.show()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and linear regression model 
# for mouse weight and average tumor volume for the Capomulin regimen
correlation = st.pearsonr(tumor_by_weight["Weight (g)"],tumor_by_weight["Tumor Volume (mm3)"])
print(f"The correlation between both factors is {round(correlation[0],2)}")

x_values = tumor_by_weight["Weight (g)"]
y_values = tumor_by_weight["Tumor Volume (mm3)"]
(slope, intercept, rvalue, pvalue, stderr) = st.linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.scatter(x_values,y_values)
plt.plot(x_values,regress_values,"r-")
plt.annotate(line_eq,(20,38),fontsize=15,color="red")
plt.xlabel('Weight (g)')
plt.ylabel('Tumor Volume (mm3)')

plt.show()
```
